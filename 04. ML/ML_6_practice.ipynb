```python
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
```

# Загрузка данных

```python
data = pd.read_excel('../data/data_ford_price.xlsx') 
```

## Предобработка данных

```python
data = data[['price','year', 'cylinders', 'odometer', 'lat', 'long', 'weather']]
data.dropna(inplace = True)

y = data['price']
x = data.drop(columns='price')
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=40)
```

#  Отбор признаков: классификация методов

## Метод рекурсивного исключения признаков

```python
from sklearn.feature_selection import RFE
```

```python
estimator = LinearRegression()
selector = RFE(estimator, n_features_to_select=3, step=1)
selector = selector.fit(X_train, y_train)
```

```python
cols_to_del = list(selector.get_feature_names_out())

cols_to_del.append('price')
print(cols_to_del)
x = data.drop(columns=cols_to_del)
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=40)
X_train.columns
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
y_predicted = model.predict(X_test)
 
mae = mean_absolute_error(y_test, y_predicted)
print('MAE: %.3f' % mae)
```

##  МЕТОДЫ ВЫБОРА ПРИЗНАКОВ НА ОСНОВЕ ФИЛЬТРОВ

```python
from sklearn.feature_selection import SelectKBest, f_regression
```

```python
y = data['price']
x = data.drop(columns='price')
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=40)
```

```python
selector = SelectKBest(f_regression, k=3)
selector.fit(X_train, y_train)
selector.get_feature_names_out()
```

```python
cols_to_del = list(selector.get_feature_names_out())
cols_to_del.append('price')
print(cols_to_del)
x = data.drop(columns=cols_to_del)
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=40)
X_train.columns
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
y_predicted = model.predict(X_test)
 
mae = mean_absolute_error(y_test, y_predicted)
print('MAE: %.3f' % mae)
```

### Выводы: 
#### метод рекурсивного исключения показал большую эффективность на нашем датасете в сравнении с методом на основе фильтров.
