```python
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
```

# Загрузка данных

```python
data = pd.read_excel('../data/data_ford_price.xlsx') 
```

## Предобработка данных

```python
data = data[['price','year', 'cylinders', 'odometer', 'lat', 'long', 'weather']]
data.dropna(inplace = True)

y = data['price']
x = data.drop(columns='price')
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=40)
```

#  Отбор признаков: классификация методов

## Метод рекурсивного исключения признаков

```python
from sklearn.feature_selection import RFE
```

```python
# Make a filtering of features by using RFE algorithm
estimator = LinearRegression()
selector = RFE(estimator, n_features_to_select=3, step=1)
selector = selector.fit(X_train, y_train)
```

```python
# Get the best features for modeling
selecled_features = list(selector.get_feature_names_out())
print(selecled_features)

# Generate a new data set for modeling
x = data[selecled_features]
```

```python
# Split the data set
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=40)
X_train.columns
```

```python
# Modeling routins
model = LinearRegression()
model.fit(X_train, y_train)
y_predicted = model.predict(X_test)

# Value a needed metric 
mae = mean_absolute_error(y_test, y_predicted)
print('MAE: %.3f' % mae)
```

##  МЕТОДЫ ВЫБОРА ПРИЗНАКОВ НА ОСНОВЕ ФИЛЬТРОВ

```python
from sklearn.feature_selection import SelectKBest, f_regression
```

```python
# Reload unmodified set
y = data['price']
x = data.drop(columns='price')
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=40)
```

```python
# Make a filtering of features by using K-best algorithm
selector = SelectKBest(f_regression, k=3)
selector.fit(X_train, y_train)
selector.get_feature_names_out()
```

```python
# Get the best features for modeling
selecled_features = list(selector.get_feature_names_out())
print(selecled_features)

# Generate a new data set for modeling
x = data[selecled_features]
```

```python
# Split the data set
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=40)
X_train.columns
```

```python
# Modeling routins
model = LinearRegression()
model.fit(X_train, y_train)
y_predicted = model.predict(X_test)

# Value a needed metric  
mae = mean_absolute_error(y_test, y_predicted)
print('MAE: %.3f' % mae)
```

### Выводы: 
#### метод рекурсивного исключения показал меньшую эффективность на нашем датасете в сравнении с методом на основе фильтров.
