```python
fig, axes = plt.subplots(nrows=3, ncols=1, figsize=(13, 9))
boxplot1 = sns.boxplot(
    data=data,
    x='age',
    orient='h',
    ax=axes[0],
    width=0.9
)
boxplot2 = sns.boxplot(
    data=data,
    x='bmi',
    # y='Exited',
    orient='h',
    ax=axes[1],
    width=0.9
);
boxplot3 = sns.boxplot(
    data=data,
    x='charges',
    y='children',
    orient='h',
    ax=axes[2],
    width=0.9
)

boxplot1.set_ylabel('age');

boxplot2.set_ylabel('bmi');

boxplot3.set_xlabel('charges');
boxplot3.set_ylabel('children');

boxplot1.grid()
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(15, 15))
barplot1 = sns.barplot(
    data=data,
    x='sex',
    y='charges',
    ax=axes[0, 0]
)
barplot2 = sns.barplot(
    data=data,
    x='children',
    y='charges',
    ax=axes[0, 1]
);
barplot3 = sns.barplot(
    data=data,
    x='smoker',
    y='charges',
    ax=axes[1, 0]
)
barplot4 = sns.barplot(
    data=data,
    x='region',
    y='charges',
    ax=axes[1, 1]
);
barplot1.set_xlabel('sex');
barplot1.set_ylabel('charges');

barplot2.set_xlabel('children');
barplot2.set_ylabel('charges');

boxplot3.set_xlabel('smoker');
barplot3.set_ylabel('charges');

barplot4.set_xlabel('region');
barplot4.set_ylabel('charges');

barplot1.grid()
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(20, 20))
fig1 = sns.scatterplot(
    data=data,
    x='age', 
    y='charges',
    hue='children',
    s=100,
    size='sex',
    ax=axes[0, 0]
)
fig2 = sns.scatterplot(
    data=data,
    x='age', 
    y='charges',
    hue='smoker',
    s=100,
    size='sex',
    ax=axes[0, 1]
)
fig3 = sns.scatterplot(
    data=data,
    x='bmi', 
    y='charges',
    hue='children',
    s=100,
    size='sex',
    ax=axes[1, 0]
)
fig4 = sns.scatterplot(
    data=data,
    x='bmi', 
    y='charges',
    hue='smoker',
    s=100,
    size='sex',
    ax=axes[1, 1]
)
fig.show()
```
